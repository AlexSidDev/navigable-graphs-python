```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('recalls_baseline.txt', 'r') as fout:
    recalls = eval(fout.read())
with open('avg_cals_baseline.txt', 'r') as fout:
    avg_cals = eval(fout.read())

fig, axes = plt.subplots(1, 2, figsize=(25, 5))
efs = list(range(5, 50, 1))

axes[0].plot(efs, recalls)
axes[0].set_xlabel('EF')
axes[0].set_ylabel('Recall')

axes[1].plot(efs, avg_cals)
axes[1].set_xlabel('EF')
axes[1].set_ylabel('Avg Calls')

with open('recalls_my_recursive_3_prune.txt', 'r') as fout:
    recalls = eval(fout.read())
with open('avg_cals_my_recursive_3_prune.txt', 'r') as fout:
    avg_cals = eval(fout.read())


axes[0].plot(efs, recalls)

axes[1].plot(efs, avg_cals)

axes[0].legend(['Baseline', 'Modified'])
axes[1].legend(['Baseline', 'Modified'])
```
